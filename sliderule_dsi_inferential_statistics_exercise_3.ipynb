# Hospital Readmissions Data Analysis and Recommendations for Reduction

### Background
In October 2012, the US government's Center for Medicare and Medicaid Services (CMS) began reducing Medicare payments for Inpatient Prospective Payment System hospitals with excess readmissions. Excess readmissions are measured by a ratio, by dividing a hospital’s number of “predicted” 30-day readmissions for heart attack, heart failure, and pneumonia by the number that would be “expected,” based on an average hospital with similar patients. A ratio greater than 1 indicates excess readmissions.

Null Hypothesis: The readmission rates between the small and large discharage are the same 

Alternative Hyptheis: The readmission rates between the small and large discharage are not the same. Significant level = 0.01

```python
def permutate_replicate_1d(data1,data2,n):
    
    # bs_sample1=np.random.choice(data1,n)
    # bs_sample2=np.random.choice(data2,n)
    
    # Concatenate the data sets: data
    data = np.concatenate((data1, data2))

    # Permute the concatenated array: permuted_data
    permuted_data = np.random.permutation(data)

    # Split the permuted array into two: perm_sample_1, perm_sample_2
    perm_sample_1 = permuted_data[:len(data1)]
    perm_sample_2 = permuted_data[len(data1):]

    return perm_sample_1,perm_sample_2

    # return bs_sample1,bs_sample2

def draw_permutate_reps(data1,data2,func,size=1):
    # bs_replicates=np.empty(size)
    
    bs_diff=np.empty(size)
    
    for i in range (size):
        bs_sample1,bs_sample2=permutate_replicate_1d(data1,data2,10000)
        bs_diff[i]=func(bs_sample1,bs_sample2)
        
    return bs_diff


def mean_diff(data_1, data_2):
   
    # The difference of means of data_1, data_2: diff
    diff = np.mean(data_1) - np.mean(data_2)

    return diff

# Filtered data into small discharge with either excess admission or small admission and large discharge with either
# excess admission or small admission
sd=data[(data['Number of Discharges']<=300)]
ld=data[(data['Number of Discharges']>300)]
plt.hist(sd['Excess Readmission Ratio'], bins=100, normed=True)

plt.figure()
plt.hist(ld['Excess Readmission Ratio'], bins=100, normed=True)
plt.show()

org_diff_mean=mean_diff(sd['Excess Readmission Ratio'],ld['Excess Readmission Ratio'])
bs_diff=draw_permutate_reps(sd['Excess Readmission Ratio'],ld['Excess Readmission Ratio'],mean_diff,10000)

# Compute p-value: p
p = np.sum(abs(bs_diff) >= abs(org_diff_mean)) / len(bs_diff)

print ('The observed difference', org_diff_mean)
print ('P-Value', p)
```
